```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()      # numeric data will show only 
```

```python
df.info()   # show data types
```

```python
df.isnull().sum()       # showing how many value null
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"] = df ["WklyStudyHours"].str.replace("5 - 10","5 - 10")
df.head()
```

# gender distribution

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from the above chart we have analysed that:
# the number of female in the data more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot = True)
plt.title("Relationship Between Parent's Education and Student's Score")
plt.show()
```

```python
# form above chart we have concluded that the education of the parents have a good impact on their study
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot = True)
plt.title("Relationship Between Parent's Marital Status and Student's Score")
plt.show()
```

```python
# from the above chart we have concluded that there is no/negligiable impact on the student score due to their parents Marital Status
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mylist = groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"],groupD["EthnicGroup"]

plt.pie(mylist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
