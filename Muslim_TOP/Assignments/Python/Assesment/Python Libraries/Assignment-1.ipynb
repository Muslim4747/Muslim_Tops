```python
import pandas as pd
```

```python
# 1) Import pandas and read the CSV file into a dataframe
banks = pd.read_csv("/content/sample_data/banklist.csv")
```

```python
# 2) Show the head of the dataframe
print(banks.head())
```

```python
# 3) What are the column names?
print(banks.columns)
```

```python
# 4) How many States (ST) are represented in this data set?
num_states = banks["ST"].nunique()
print(f"Number of unique states: {num_states}")
```

```python
# 5) Get a list or array of all the states in the data set
states_list = banks["ST"].unique()
print(f"List of all states: {states_list}")
```

```python
# 6) What are the top 5 states with the most failed banks?
top_states = banks["ST"].value_counts().head(5)
print("Top 5 states with the most failed banks:\n", top_states)
```

```python
#7) What are the top 5 acquiring institutions?
top_acquirers = banks["Acquiring Institution"].value_counts().head(5)
print("Top 5 acquiring institutions:\n", top_acquirers)
```

```python
# 8) How many banks has the State Bank of Texas acquired? How many were actually in Texas?
state_bank_acquired = banks[banks["Acquiring Institution"] == "State Bank of Texas"]
num_acquired = len(state_bank_acquired)
num_in_texas = len(state_bank_acquired[state_bank_acquired["ST"] == "TX"])
print(f"State Bank of Texas acquired {num_acquired} banks, of which {num_in_texas} were in Texas.")
```

```python
# 9) What is the most common city in California for a bank to fail in?
california_banks = banks[banks["ST"] == "CA"]
most_common_city = california_banks["City"].value_counts().idxmax()
print(f"The most common city in California for a bank to fail in is: {most_common_city}")
```
